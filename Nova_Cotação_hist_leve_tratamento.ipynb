```python
import pandas as pd
import yfinance as yf
import os
```

```python
# Usando arquivo CSV com lista de empresas baixado do site da B3 #

listacsv = pd.read_csv('acoeslistadasb3_26062025.csv')
empresas = []
#listacsv

for nome in listacsv['Ticker']:
    empresas.append(nome)

print (empresas)
```

```python
# Removendo códigos com final 11 

for nome in empresas:
    tam=len(nome)
    if tam>5:
        empresas.remove(nome)
empresas.sort()
qtd = len(empresas)
print(qtd,empresas)
```

```python
# Acrescentar ".SA" nos tickers das empresas - Padrão YFinance de busca

ticker = []
for empresa in empresas:
    empresa = f'{empresa}.SA'
    ticker.append(empresa)

print (ticker)
    
```

# Pegar cotação histórica (Apenas o preço de fechamento "Close")

```python

inicio = "2010-01-02"
fim = "2024-07-01"

cotacao_df=yf.download(tickers=ticker, start=inicio, end=fim)["Close"]
    

display(cotacao_df)
```

```python
cotacao_df.to_excel('cotacoes.xlsx')
```

```python
# Remover ".SA" dos nomes das empresas nas colunas da tabela
df = pd.read_excel("cotacoes.xlsx")

cotacoes_df = df
#display(cotacoes_df)
for empresa in cotacoes_df.columns:
    if empresa == "Date":
        pass
    
    else:
    
        novonome = empresa[-8:-3]

        #print(novonome)
        cotacoes_df.rename(columns={f'{empresa}': f'{novonome}'}, inplace=True)

cotacoes_df = cotacoes_df.set_index('Date') 

display(cotacoes_df) 
```

```python
# Separar as cotações em tabelas de cada empresa 

for empresa in cotacoes_df:
    if empresa == 'Date':
        pass
    else:
        tabela = cotacoes_df[empresa]
       
        cotacao = tabela.to_excel(f'./cotacoes/{empresa}.xlsx')
```
